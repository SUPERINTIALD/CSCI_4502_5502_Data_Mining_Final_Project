# ETS Decomposition
Used to separate different components of a time series. The term ETS stands for Error, Trend, and Seasonality.

1. Error (E): Represents the residuals or noise in the data—variations that can't be attributed to any specific pattern.
2. Trend (T): Reflects the underlying direction in the data over time (e.g., a general increase or decrease in housing prices).
3. Seasonal (S): Captures repeating patterns or cycles that recur over regular intervals (e.g., seasonal fluctuations in prices across months).

```python
# Convert data column to be datatime
boulder_df_house['Date'] = pd.to_datetime(boulder_df_house['Date'])  # Convert to datetime if not already
boulder_df_house.set_index('Date', inplace=True)

# Set frequency to monthly
boulder_df_house = boulder_df_house.asfreq('M')
```
